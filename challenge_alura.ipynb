### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Calcular la facturación (Precio + Costo de envío)
tienda['Facturación'] = tienda['Precio'] + tienda['Costo de envío']
tienda2['Facturación'] = tienda2['Precio'] + tienda2['Costo de envío']
tienda3['Facturación'] = tienda3['Precio'] + tienda3['Costo de envío']
tienda4['Facturación'] = tienda4['Precio'] + tienda4['Costo de envío']

# Agrupar por tienda (Lugar de Compra) y sumar la facturación
facturacion_por_tienda = tienda.groupby('Lugar de Compra')['Facturación'].sum().reset_index()
facturacion_por_tienda2 = tienda2.groupby('Lugar de Compra')['Facturación'].sum().reset_index()
facturacion_por_tienda3 = tienda3.groupby('Lugar de Compra')['Facturación'].sum().reset_index()
facturacion_por_tienda4 = tienda4.groupby('Lugar de Compra')['Facturación'].sum().reset_index()
```

```python
# Mostrar resultados
print("Facturación total por tienda:\n")
print(facturacion_por_tienda)

print("Facturación total por tienda2:\n")
print(facturacion_por_tienda2)

print("Facturación total por tienda3:\n")
print(facturacion_por_tienda3)

print("Facturación total por tienda4:\n")
print(facturacion_por_tienda4)
```

# 2. Ventas por categoría

```python
ventas_por_categoria = tienda.groupby('Categoría del Producto')['Precio'].sum().reset_index().sort_values('Precio', ascending=False)
ventas_por_categoria2 = tienda2.groupby('Categoría del Producto')['Precio'].sum().reset_index().sort_values('Precio', ascending=False)
ventas_por_categoria3 = tienda3.groupby('Categoría del Producto')['Precio'].sum().reset_index().sort_values('Precio', ascending=False)
ventas_por_categoria4 = tienda4.groupby('Categoría del Producto')['Precio'].sum().reset_index().sort_values('Precio', ascending=False)
```

```python
print("Las ventas por categoría tienda 1: \n", ventas_por_categoria)
print("Las ventas por categoría tienda 2: \n", ventas_por_categoria2)
print("Las ventas por categoría tienda 3: \n", ventas_por_categoria3)
print("Las ventas por categoría tienda 4: \n", ventas_por_categoria4)
```

# 3. Calificación promedio de la tienda




```python
# Calcular la calificación promedio
calificacion_promedio = tienda['Calificación'].mean()
calificacion_promedio2 = tienda2['Calificación'].mean()
calificacion_promedio3 = tienda3['Calificación'].mean()
calificacion_promedio4 = tienda4['Calificación'].mean()
```

```python
#Mostrar resultado
print(f'La calificación promedio de la tienda es: {calificacion_promedio:.2f}')
print(f'La calificación promedio de la tienda 2 es: {calificacion_promedio2:.2f}')
print(f'La calificación promedio de la tienda 3 es: {calificacion_promedio3:.2f}')
print(f'La calificación promedio de la tienda 4 es: {calificacion_promedio4:.2f}')
```
